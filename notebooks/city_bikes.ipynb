```python
import pandas as pd
import requests 
```

# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
api_url = f'http://api.citybik.es/v2/networks/citybikelinz'
response = requests.get(api_url)
data = response.json()
response.status_code
data
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
stations = data['network']['stations']

stations_data = []
for station in stations:
    stations_name = station['name']
    latitude = station['latitude']
    longitude = station['longitude']
    num_bikes = station['free_bikes']
    stations_data.append({'stations_name': stations_name, 'latitude': latitude, 'longitude': longitude, 'number of bikes': num_bikes})

```

Put your parsed results into a DataFrame.

```python
df_citybike = pd.DataFrame(stations_data)
df_citybike
```
